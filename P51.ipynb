```python
import scipy.integrate as integrate
import math
import matplotlib.pyplot as plt
import numpy as np

Tfun = lambda x: x - (x**3)/math.factorial(3) +(x**5)/math.factorial(5)
fun = lambda x: (math.sin(x) - Tfun(x))**2
errorFun = lambda x: math.sqrt(integrate.quad(fun,0,x)[1])
npoints = 100
xvec = np.linspace(0,1,npoints)

analytical_values = [errorFun(i) for i in xvec]
plt.plot(xvec, analytical_values)
plt.show()
```

```python
import MomentFitP49

Tfun = lambda x: x - (x**3)/math.factorial(3) +(x**5)/math.factorial(5)

efun = lambda x, quadrule : abs(integrate.quad(math.sin,0,x)[1]-MomentFitP49.GauseLegendreQuadrature(Tfun,quadrule))   

quadrules = range(1,4)

A=np.empty([3,npoints])

for i in range(3):
    for j in range(npoints):
        A[i][j] = efun(xvec[j],quadrules[i])

plt.plot(xvec, A[0])
plt.plot(xvec, A[1])
plt.plot(xvec, A[2])
plt.show()
```
